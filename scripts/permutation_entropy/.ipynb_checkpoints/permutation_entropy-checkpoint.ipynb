```python
import numpy as np
import ordpy
import matplotlib.pyplot as plt
from glob import glob
import os
import scipy.io as sio
import nibabel as nib
```

```python
MAIN_DIR = "/home/usuario/disco1/proyectos/2023-resting-state-estados-fMRI_conn"
N_rois = 264
subs = glob(f"{MAIN_DIR}/results/func_networks/consensus_264/*")
N_subs = len(subs)
brain_mask = sio.loadmat(f"{MAIN_DIR}/scripts/seed/masks/brainmask.mat")["mask"]
print(brain_mask.shape)
```

```python
#por red funcional o ROI
he_matrix = np.zeros((N_subs, 4, N_rois)) #subs x conditions
sc_matrix = np.zeros((N_subs, 4, N_rois)) #subs x conditions
for iSub, sub in enumerate(subs):
    print(f"subject {os.path.basename(sub)}")
    for iCond in [0,1,2,3]:
        data = sio.loadmat(f"{sub}/cond{iCond+1}/func_roi.mat")["func_roi"]
        for iRoi in range(N_rois):
            he, sc = ordpy.complexity_entropy(data[iRoi,:], dx=4)
            he_matrix[iSub, iCond, iRoi] = he
            sc_matrix[iSub, iCond, iRoi] = sc
        
```

```python
roi_id = np.loadtxt(f"{MAIN_DIR}/scripts/func_networks/masks/rois264_identity.txt")
he_func_network = np.zeros((N_subs, 4, 14)) #subjects x condition x number of func networks
sc_func_network = np.zeros((N_subs, 4, 14)) #subjects x condition x number of func networks
for iFunc in range(1,15):
    roi_list = roi_id == iFunc
    he_func_network[:, :, iFunc-1] = np.mean(he_matrix[:, :, roi_list], axis=2)
    sc_func_network[:, :, iFunc-1] = np.mean(sc_matrix[:, :, roi_list], axis=2)
```

## figures

```python
he_mean = np.mean(he_matrix, axis=0)
he_stderr = np.std(he_matrix, axis=0)/np.sqrt(N_subs)
plt.bar([1,2,3,4], he_mean, yerr=he_stderr)
plt.ylim([0.80, 0.81])
plt.title("Permutation entropy")
plt.xticks([1,2,3,4], ["rep", "alt", "trans", "recu"])
```

```python
sc_mean = np.mean(sc_matrix, axis=0)
sc_stderr = np.std(he_matrix, axis=0)/np.sqrt(N_subs)
plt.bar([1,2,3,4], sc_mean, yerr=sc_stderr)
plt.ylim([0.2225, 0.2350])
plt.title("Statistical complexity")
plt.xticks([1,2,3,4], ["rep", "alt", "trans", "recu"])
```

```python
offset_x = 0.00005
plt.scatter(he_mean, sc_mean)
plt.annotate("reposo", (he_mean[0]+offset_x, sc_mean[0]))
plt.annotate("transicion", (he_mean[1]+offset_x, sc_mean[1]))
plt.annotate("alteracion", (he_mean[2]+offset_x, sc_mean[2]))
plt.annotate("recuperacion", (he_mean[3]+offset_x, sc_mean[3]))
plt.ylabel("statistical complexity")
plt.xlabel("permutation entropy")
plt.xlim([0.803, 0.806])
plt.ylim([0.2295, 0.2315])
```

```python
network_names = ["UNC", "SSH", "SSM", "COTC", "AUD", "DMN", "MEM", "VA", "VIS", "FPC", "SA", "SUB", "CER", "DA", "OTRO"]
fig, axs = plt.subplots(14,1)
fig.set_size_inches(7, 70)
for iFunc in range(14):
    means = np.mean(he_func_network[:,:,iFunc], axis=0)
    stderrs = np.std(he_func_network[:,:,iFunc], axis=0)/np.sqrt(N_subs)
    axs[iFunc].bar([1,2,3,4], means, yerr=stderrs)
    axs[iFunc].set_ylim([np.min(means)-1.5*np.max(stderrs), np.max(means)+1.5*np.max(stderrs)])
    axs[iFunc].set_title(network_names[iFunc])
    axs[iFunc].set_ylabel("permutation entropy")
    axs[iFunc].set_xticks([1,2,3,4])
    axs[iFunc].set_xticklabels(["reposo", "transicion", "alteracion", "recuperacion"])
```

```python
network_names = ["UNC", "SSH", "SSM", "COTC", "AUD", "DMN", "MEM", "VA", "VIS", "FPC", "SA", "SUB", "CER", "DA", "OTRO"]
fig, axs = plt.subplots(14,1)
fig.set_size_inches(7, 70)
for iFunc in range(14):
    means = np.mean(sc_func_network[:,:,iFunc], axis=0)
    stderrs = np.std(sc_func_network[:,:,iFunc], axis=0)/np.sqrt(N_subs)
    axs[iFunc].bar([1,2,3,4], means, yerr=stderrs)
    axs[iFunc].set_ylim([np.min(means)-1.5*np.max(stderrs), np.max(means)+1.5*np.max(stderrs)])
    axs[iFunc].set_title(network_names[iFunc])
    axs[iFunc].set_ylabel("statistical complexity")
    axs[iFunc].set_xticks([1,2,3,4])
    axs[iFunc].set_xticklabels(["reposo", "transicion", "alteracion", "recuperacion"])
```
